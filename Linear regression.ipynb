```python
from sklearn import linear_model
import numpy as np
import pandas as pd
# use pandas to input training data file and test data file, no header
# transform arrry to dataframe
traindata=pd.DataFrame(pd.read_excel('/Users/nicole_he_dm/Documents/dartmouth study/Machine learning and statistical analysis/homework/homework4 training.xlsx',header=None))
testdata=pd.DataFrame(pd.read_excel('/Users/nicole_he_dm/Documents/dartmouth study/Machine learning and statistical analysis/homework/homework4 test.xlsx',header=None))
traindata.head()
```

```python
X_train = traindata[[1,2,3,4,5,6,7]]
# X_train is last 7 fields of independent variables in traindata
X_train.head()
```

```python
# replace '?' with mean value of each column
X_train=X_train.replace('?',np.NaN)
X_train=X_train.fillna(X_train.mean())
# Y_train is the list of dependent variable in the traindata
Y_train=traindata[[0]]
Y_train.head()
```

```python
# train the model
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
linreg.fit(X_train, Y_train)
```
